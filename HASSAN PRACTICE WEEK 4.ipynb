```python
Index['Age Category'] = ['iGen' if 0<age<25 else 'Millenials' if 24<age<40 else 'GenX' if 39<age<56 else 'BabyBoomers' if 55<age<74 else None \
 for age in list(Index['Age'].values)]
```

```python
Index.head()
```

```python
Index['Income Category'] = pd.qcut(Index['EstimatedSalary'],3,labels=['Low','Medium','High'])
```

```python
Index.head()
```

```python
Index['Income Category'] = pd.qcut(Index['EstimatedSalary'],4,labels=['Low','LowMedium' ,'Medium','High'])
```

```python
Index.head()
```

```python
age_cat = Index.groupby('Age Category').count()
inc_cat = Index.groupby('Income Category').count()
```

```python
plt.bar(age_cat.index.values, age_cat['Age'], color = 'orange')
plt.ylabel("Numbers of Users")
plt.xlabel("Age Category")
plt.show()
```

```python
plt.bar(inc_cat.index.values, inc_cat['Age'], color = 'grey')
plt.ylabel("Numbers of Users")
plt.xlabel("Income Category")
plt.show()
```

```python
features= ['Gender', 'Age Category', 'Income Category']
fig= plt.subplots(figsize= (10,15))

for i,j in enumerate(features):
    plt.subplot(4,2, i+1)
    plt.subplots_adjust(hspace=1.0)
    sns.countplot(x=j, Index=Index , hue= "Purchased")
    plt.xticks(rotation= 90)
    plt.title("No of Employee")
```
